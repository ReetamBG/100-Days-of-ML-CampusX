```python
import numpy as np
import pandas as pd
```

```python
date = pd.read_csv('./orders.csv')
time = pd.read_csv('./messages.csv')
```

```python
date.head()
```

```python
date['date'] = pd.to_datetime(date['date'])
date.info()
```

```python
date['year'] = date['date'].dt.year
date.head()
```

```python
date['month'] = date['date'].dt.month
date.head()
```

```python
date['month_name'] = date['date'].dt.month_name()
date.head()
```

```python
date['day'] = date['date'].dt.day
date.head()
```

```python
date['day_of_week'] = date['date'].dt.dayofweek
date.head()
```

```python
date['day_name'] = date['date'].dt.day_name()
date.head()
```

```python
date['is_weekend'] = np.where(date['day_name'].isin(['Saturday', 'Sunday']), 1, 0)
date.head()
```

```python
date['semester'] = np.where(date['date'].dt.quarter.isin([1,2]), 1, 2)
date.sample(5)
```

```python
date['quarter'] = date['date'].dt.quarter
date.sample(5)
```
