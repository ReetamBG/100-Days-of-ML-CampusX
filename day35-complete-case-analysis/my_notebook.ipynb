```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./data_science_job.csv')
df.head()
```

```python
# (df.isnull().sum()/df.shape[0]) * 100
df.isnull().mean() * 100
```

```python
cols = [col for col in df.columns if df[col].isnull().mean()*100<5 and df[col].isnull().mean()*100>0]
cols
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(15, 3))

col_list = ['city_development_index', 'experience', 'training_hours']

for col,i in zip(col_list, range(3)):
    sns.histplot(data=df, x=col, ax=ax[i], kde=True) 
```

```python
new_df = df[cols].dropna()
new_df.shape, df.shape
```

```python
len(new_df)/len(df)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

col_list = ['city_development_index', 'experience', 'training_hours']

for col,i in zip(col_list, range(3)):
    sns.histplot(data=df, x=col, ax=ax[i], alpha=0.7) 
    sns.histplot(data=new_df, x=col, ax=ax[i])     
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

col_list = ['city_development_index', 'experience', 'training_hours']

for col,i in zip(col_list, range(3)):
    sns.kdeplot(data=df, x=col, ax=ax[i]) 
    sns.kdeplot(data=new_df, x=col, ax=ax[i])   
```

```python
col_list = ['education_level', 'enrolled_university']

ratio_before_cca = df[col_list[0]].value_counts() / df.shape[0]
ratio_after_cca = new_df[col_list[0]].value_counts() / new_df.shape[0]

ratio_comp_1 = pd.concat([ratio_before_cca, ratio_after_cca], axis=1)
ratio_comp_1.columns = ['before cca', 'after cca']
ratio_comp_1
```
