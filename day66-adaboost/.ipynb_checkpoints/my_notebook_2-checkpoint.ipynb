# Adaboost Hyperparameters

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import AdaBoostClassifier, AdaBoostRegressor
```

```python
# plotting functions

def plot_decision_regions(X, y, model):
    cv_score = cross_val_score(estimator=model, X=X, y=y)
    print('Mean Cross Validation Score :', np.mean(cv_score))
    
    x_ = np.linspace(min(X[:,0])-0.2, max(X[:,0])+0.2, 300)
    y_ = np.linspace(min(X[:,1])-0.2, max(X[:,1])+0.2, 300)
    xx, yy = np.meshgrid(x_, y_)
    model_input = np.concat((xx.reshape(-1,1), yy.reshape(-1,1)), axis=1)
    model.fit(X, y)
    z = model.predict(model_input)
    zz = z.reshape(xx.shape)
    plt.contourf(xx, yy, zz, alpha=0.2)
    plt.scatter(X[:,0], X[:,1], c=y, alpha=0.7)
    plt.show()

def plot_regression_line(X, y, model):
    cv_score = cross_val_score(estimator=model, X=X, y=y)
    print('Mean Cross Validation Score :', np.mean(cv_score))
    
    X_plot = np.linspace(min(X)-1, max(X)+1, 100)
    model.fit(X, y)
    y_plot = model.predict(X_plot)
    plt.plot(X_plot, y_plot)
    plt.scatter(X, y)
    plt.show()
```

## Classification

```python
from sklearn.datasets import make_circles

X, y = make_circles(n_samples=300, noise=0.2, factor=0.5, random_state=42)
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
# Using decision tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

dtc = DecisionTreeClassifier(max_depth=None)
plot_decision_regions(X, y, dtc)
```

#### We can see quite a lot of overfitting 

```python
# Using AdaBoostClassifier
from sklearn.ensemble import AdaBoostClassifier

abc = AdaBoostClassifier(algorithm='SAMME')
plot_decision_regions(X, y, abc)
```

#### We can see that it becomes a low bias low variance model

```python
# only one weak estimator (decision stump)- high bias low variance
abc = AdaBoostClassifier(n_estimators=1, algorithm='SAMME')
plot_decision_regions(X, y, abc)
```

```python
# inreasing n_estimators increases fit
abc = AdaBoostClassifier(n_estimators=1000, algorithm='SAMME')
plot_decision_regions(X, y, abc)
```

```python
# Increasing n_estimators while reducing learning rate can compensate for the overfitting
abc = AdaBoostClassifier(n_estimators=1000, learning_rate=0.1, algorithm='SAMME')
plot_decision_regions(X, y, abc)
```

### Using other estimators

```python
# fully grown decision tree
abc = AdaBoostClassifier(estimator=DecisionTreeClassifier(max_depth=None), n_estimators=100, learning_rate=0.1, algorithm='SAMME')
plot_decision_regions(X, y, abc)
```

```python
# logistic regression
from sklearn.linear_model import LogisticRegression
abc = AdaBoostClassifier(estimator=LogisticRegression(), n_estimators=100, learning_rate=0.1, algorithm='SAMME')
plot_decision_regions(X, y, abc)
```

```python
# svm
from sklearn.svm import SVC

svc = SVC()
plot_decision_regions(X, y, svc)
```

```python
abc = AdaBoostClassifier(estimator=SVC(), n_estimators=100, algorithm='SAMME')
plot_decision_regions(X, y, abc)
```
