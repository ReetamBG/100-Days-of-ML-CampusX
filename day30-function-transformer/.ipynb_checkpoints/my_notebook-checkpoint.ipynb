```python
import numpy as np
import pandas as pd
import scipy.stats as stats

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import FunctionTransformer
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split, cross_val_score

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('./train.csv', usecols=['Age', 'Fare', 'Survived'])
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
# splitting the data

X_train, X_test, y_train, y_test = train_test_split(df[['Age', 'Fare']], df['Survived'], test_size=0.25, random_state=42)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,3))

sns.kdeplot(data=X_train, x='Age', ax=ax[0])
ax[0].set_title('Age Distribution')

sns.kdeplot(data=X_train, x='Fare', ax=ax[1])
ax[1].set_title('Fare Distribution')

plt.show()
```

### Age is slightly right skewed but almost normal
### Fare is highly skewed to the right

```python
plt.figure(figsize=(10,3))

plt.subplot(121)
stats.probplot(X_train['Age'], dist='norm', plot=plt)
plt.title('Age')

plt.subplot(122)
stats.probplot(X_train['Fare'], dist='norm', plot=plt)
plt.title('Fare')

plt.show()
```

```python
ft = FunctionTransformer(func=np.log1p)
```

```python
X_train_transformed = ft.fit_transform(X_train)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,3))

sns.kdeplot(data=X_train_transformed, x='Age', ax=ax[0])
ax[0].set_title('Age distribution after transformation')

sns.kdeplot(data=X_train_transformed, x='Fare', ax=ax[1])
ax[1].set_title('Fare distribution after transformation')
```

```python
plt.figure(figsize=(10,3))

plt.subplot(121)
stats.probplot(X_train_transformed['Age'], dist='norm', plot=plt)
plt.title('Age distribution after function transformation')

plt.subplot(122)
stats.probplot(X_train_transformed['Fare'], dist='norm', plot=plt)
plt.title('Fare distribution after function transformation')

plt.show()
```

```python
# predictions after transformation

plt.figure(figsize=(10,5))

plt.subplot(221)
sns.kdeplot(X_train['Age'])
plt.title('Age before')

plt.subplot(222)
sns.kdeplot(X_train['Fare'])
plt.title('Fare before')

ft = FunctionTransformer(lambda x : x**2)
plt.subplot(223)
sns.kdeplot(ft.fit_transform(X_train['Age']))
plt.title('Age after')

print(f'''
Age before root : {X_train['Age'].skew()}
Age after root : {ft.fit_transform(X_train['Age']).skew()}
''')

ft = FunctionTransformer(np.log1p)
plt.subplot(224)
sns.kdeplot(ft.fit_transform(X_train['Fare']))
plt.title('Fare after')

print(f'''
Fare before log : {X_train['Age'].skew()}
Fare after log : {ft.fit_transform(X_train['Age']).skew()}
''')
```

```python
# predictions before transformation

np.random.seed(42)

clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()

clf1.fit(X_train, y_train)
clf2.fit(X_train, y_train)

y_pred_1 = clf1.predict(X_test)
y_pred_2 = clf2.predict(X_test)

acc_1 = accuracy_score(y_test, y_pred_1)
acc_2 = accuracy_score(y_test, y_pred_2)

print(f'''
Accuracy
Logistic regression : {acc_1}
Decision tree classifier : {acc_2}
''')
```

```python
# after transformation

ct = ColumnTransformer(transformers=[
        ('age_transformer', FunctionTransformer(None), ['Age']),
        ('fare_transformer', FunctionTransformer(np.log1p), ['Fare']),
    ], remainder='passthrough')

X_train_tranformed = ct.fit_transform(X_train)
X_test_transformed = ct.transform(X_test)

clf1.fit(X_train_transformed, y_train)
clf2.fit(X_train_transformed, y_train)

y_pred_1 = clf1.predict(X_test_transformed)
y_pred_2 = clf2.predict(X_test_transformed)

acc_1 = accuracy_score(y_test, y_pred_1)
acc_2 = accuracy_score(y_test, y_pred_2)

print(f'''
Accuracy
Logistic regression : {acc_1}
Decision tree classifier : {acc_2}
''')
```

### Decision trees perform worse with transform

```python
def predict(transformer_age=None, transformer_fare=None):
    np.random.seed(42)

    ct = ColumnTransformer(transformers=[
        ('age_transformer', FunctionTransformer(transformer_age), ['Age']),
        ('fare_transformer', FunctionTransformer(transformer_fare), ['Fare']),
    ], remainder='passthrough', verbose_feature_names_out=False).set_output(transform='pandas')    # setting names and output for plotting

    X_train_transformed = ct.fit_transform(X_train)
    X_test_transformed = ct.fit_transform(X_test)

    clf1 = LogisticRegression()
    clf2 = DecisionTreeClassifier()
    
    clf1.fit(X_train_transformed, y_train)
    clf2.fit(X_train_transformed, y_train)
    
    y_pred_1 = clf1.predict(X_test_transformed)
    y_pred_2 = clf2.predict(X_test_transformed)
    
    acc_1 = accuracy_score(y_test, y_pred_1)
    acc_2 = accuracy_score(y_test, y_pred_2)
    
    print(f'''
    Accuracy
    Logistic regression : {acc_1}
    Decision tree classifier : {acc_2}
    ''')

    plt.figure(figsize=(15, 10))

    plt.subplot(221)
    stats.probplot(X_train['Age'], dist='norm', plot=plt)
    plt.title('Age before')
    plt.xlabel('')

    plt.subplot(222)
    stats.probplot(X_train_transformed['Age'], dist='norm', plot=plt)
    plt.title('Age after')
    plt.xlabel('')

    plt.subplot(223)
    stats.probplot(X_train['Fare'], dist='norm', plot=plt)
    plt.title('Fare before')

    plt.subplot(224)
    stats.probplot(X_train_transformed['Fare'], dist='norm', plot=plt)
    plt.title('Fare after')

    plt.show()
```

```python
predict(transformer_fare=np.log1p)
```

```python
predict(transformer_age=lambda x : x**2, transformer_fare=np.log1p)
```
