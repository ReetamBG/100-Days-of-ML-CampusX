```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./placement.csv')
df.head()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

df['cgpa'].plot(kind='kde', ax=ax[0])
df['placement_exam_marks'].plot(kind='kde', ax=ax[1])
plt.show()
```

```python
df['cgpa'].skew(), df['placement_exam_marks'].skew()
```

```python
sns.boxplot(data=df, x='placement_exam_marks')
```

```python
df['placement_exam_marks'].describe()
```

```python
marks = df['placement_exam_marks']

q1 = marks.quantile(0.25)
q3 = marks.quantile(0.75)
iqr = q3-q1

upper_limit = q3 + 1.5*iqr
lower_limit = q1 - 1.5*iqr

upper_limit, lower_limit
```

```python
df[(df['placement_exam_marks']>upper_limit) | (df['placement_exam_marks']<lower_limit)]
```

```python
new_df = df[(marks<upper_limit) & (marks>lower_limit)]
new_df.describe()
```

```python
sns.boxplot(data=new_df, x='placement_exam_marks')
```

```python
# capping

new_df_2 = df
new_df_2['placement_exam_marks'] = np.where(df['placement_exam_marks']>upper_limit, upper_limit,
                                            np.where(df['placement_exam_marks']<lower_limit, lower_limit,
                                                     df['placement_exam_marks']))

new_df_2.describe()
```

```python
sns.boxplot(data=new_df_2, x='placement_exam_marks')
```

```python
# plotting everything together to compare

fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(10,10))

sns.histplot(df['placement_exam_marks'], ax=ax[0,0], kde=True, bins=30)
sns.boxplot(x=df['placement_exam_marks'], ax=ax[0,1])

sns.histplot(df['placement_exam_marks'], ax=ax[1,0], kde=True, bins=30)
sns.boxplot(x=new_df['placement_exam_marks'], ax=ax[1,1])

sns.histplot(df['placement_exam_marks'], ax=ax[2,0], kde=True, bins=30)
sns.boxplot(x=new_df_2['placement_exam_marks'], ax=ax[2,1])

plt.show()
```
