```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

# Let's Plan

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
# Step 1 -> train/test/split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Survived']),
                                                 df['Survived'],
                                                 test_size=0.2,
                                                random_state=42)
```

```python
X_train.head()
```

```python
y_train.sample(5)
```

# Pipeline


```python
# imputation transformer (missing values)

trf1 = ColumnTransformer([
    ('impute_age',SimpleImputer(),[2]),   
    ('impute_embarked',SimpleImputer(strategy='most_frequent'),[6])
],remainder='passthrough')


# Note that we are using the col numbers this time. Why? Cuz transforming converts it into a np array from a dataframe
# the next transformer will recieve an np array
# dataframes have col names but np arrays dont. So indexing by numbers will work even after transformation
```

```python
# one hot encoding 

trf2 = ColumnTransformer([
    ('ohe_sex_embarked',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[1,6])    # col no. 1 and 6
],remainder='passthrough')

# -> Note that it will change the indices cuz it increases the number of columns -> next transformer in pipeline will have diff col numbers
# not using drop='first' cuz decision trees dont benefit from removing multicolinearity
```

```python
# Scaling 

trf3 = ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,10))        # col no 0 to 9 ( or do slice(0,-1))
])

# scaling columns 0 to 9 (9 cuz ohe increased number) -> categorical columns not affected ??
```

```python
# Feature selection
trf4 = SelectKBest(score_func=chi2,k=8)
```

```python
# train the model
trf5 = DecisionTreeClassifier()
```

# Create Pipeline

```python
pipe = Pipeline([
    # (name, transformer)
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    ('trf4',trf4),
    ('trf5',trf5)
])
```

# Pipeline Vs make_pipeline

Pipeline requires naming of steps, make_pipeline does not.

(Same applies to ColumnTransformer vs make_column_transformer)

```python
# Alternate Syntax
# pipe = make_pipeline(trf1,trf2,trf3,trf4,trf5)
```

```python
# train
pipe.fit(X_train,y_train)
```

# Cross Validation using Pipeline

```python
# cross validation using cross_val_score
from sklearn.model_selection import cross_val_score
cross_val_score(pipe, X_train, y_train, cv=5, scoring='accuracy').mean()
```
