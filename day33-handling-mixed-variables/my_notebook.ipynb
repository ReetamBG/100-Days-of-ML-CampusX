```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./titanic.csv')
df.head()
```

```python
df['number_num'] = pd.to_numeric(df['number'], errors='coerce')
df['number_cat'] = np.where(df['number_num'].isna(), df['number'], np.nan)
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract(r'(\d+)')
df['cabin_cat'] = df['Cabin'].str[0]
df[~df['Cabin'].isnull()].head()
```

```python
df['cabin_cat'].unique()
```

```python
sns.displot(kind='hist', data=df, x='cabin_cat', stat='probability', height=3, aspect=1.2)
plt.show()
```

```python
df['Ticket'].unique()
```

```python
df['ticket_cat'] = df['Ticket'].str.split().str[0]
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan, df['ticket_cat'])
df.head()
```

```python
df['ticket_num'] = df['Ticket'].str.split().str[-1]
df['ticket_num'] = pd.to_numeric(df['ticket_num'], errors='coerce', downcast='integer')
df.head()
```

```python
df['ticket_cat'].unique()
```

```python
sns.displot(kind='hist', data=df, x='ticket_cat', stat='probability', kde=True, height=7, aspect=3)
plt.xticks(rotation=90)
plt.show()
```
