```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
titanic = pd.read_csv('train.csv')
```

```python
flights = sns.load_dataset('flights')
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

## 1. Scatterplot (Numerical - Numerical)

```python
# bivariate
sns.scatterplot(x=tips['total_bill'], y=tips['tip'])
```

```python
# multivariate -> 5 columns in this case
sns.scatterplot(data=tips,
                x='total_bill',
                y='tip', 
                hue='sex', 
                style='smoker',
                size='size')
```

## 2. Bar Plot (Numerical - Categorical)

```python
titanic.head()
```

```python
sns.barplot(titanic['Pclass'],titanic['Age'],hue=titanic['Sex'])
```

## 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(titanic['Sex'],titanic['Age'],hue=titanic['Survived'])
```

## 4. Distplot (Numerical - Categorical)

```python
# sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
# sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)

sns.displot(data=titanic, x='Age', hue='Survived', kind='kde', fill=True)
```

## 5. HeatMap (Categorical - Categorical)

```python
titanic.head(3)
```

# Crosstab -> Good for categorical vs categorical

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
# alternate way of crosstab
titanic['Total Passengers'] = titanic['SibSp'] + titanic['Parch'] + 1
pd.pivot_table(data=titanic, index='Pclass', columns='Survived', values='Total Passengers', aggfunc='count')
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
# percentage of survival and pclass
(titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

```python
(titanic.groupby('Embarked')['Survived'].mean()*100).plot(kind='bar')
```

```python
(titanic.groupby('Sex')['Survived'].mean()*100).plot(kind='bar')
```

## 6. ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```
