```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('./train.csv', usecols=['GarageQual','FireplaceQu','SalePrice'])
df.head()
```

```python
val_counts_fq = df['FireplaceQu'].value_counts()
val_counts_fq
```

```python
val_counts_gq = df['GarageQual'].value_counts()
val_counts_gq
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

sns.barplot(data=val_counts_fq, ax=ax[0])
sns.barplot(data=val_counts_gq, ax=ax[1])

plt.show()
```

```python
# Fireplace has missing values > 5% -- not a good suitor for most freq replacement
# Moreover 2 categories are dominant -- even worse for this
# We should use arbitrary value replacement

# GarageQual has missing values < 5% -- We can use replacement with most frequent/mode
# One category (TA) is much more dominant compared to others -- Good for mode replacement
```

```python
mode_fq = df['FireplaceQu'].mode()[0]
mode_gq = df['GarageQual'].mode()[0]
mode_fq, mode_gq
```

```python
df['FireplaceQu_imputed'] = df['FireplaceQu'].fillna(mode_fq)
df['GarageQual_imputed'] = df['GarageQual'].fillna(mode_gq)

df.sample(10)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))

sns.kdeplot(data=df[df['FireplaceQu'].isnull()], x='SalePrice', ax=ax[0], label='Null')
sns.kdeplot(data=df[df['FireplaceQu']==mode_fq], x='SalePrice', ax=ax[0], label=mode_fq)

sns.kdeplot(data=df[df['GarageQual'].isnull()], x='SalePrice', ax=ax[1], label='Null')
sns.kdeplot(data=df[df['GarageQual']==mode_gq], x='SalePrice', ax=ax[1], label=mode_gq)

ax[0].legend()
ax[1].legend()
ax[0].set_title('FireplaceQu')
ax[1].set_title('GarageQual')
plt.suptitle('KDE of SalePrice for Null vs Mode')
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))

sns.kdeplot(data=df[df['FireplaceQu']==mode_fq], x='SalePrice', ax=ax[0], label='Before Imputation')
sns.kdeplot(data=df[df['FireplaceQu_imputed']==mode_fq], x='SalePrice', ax=ax[0], label='After Imputation')

sns.kdeplot(data=df[df['GarageQual']==mode_gq], x='SalePrice', ax=ax[1], label='Before Imputation')
sns.kdeplot(data=df[df['GarageQual_imputed']==mode_gq], x='SalePrice', ax=ax[1], label='After Imputation')

ax[0].legend()
ax[1].legend()
ax[0].set_title('FireplaceQu')
ax[1].set_title('GarageQual')
plt.suptitle('KDE of SalePrice for Before vs After imputation')
plt.show()
```

```python
# FOR FireplaceQu
# KDE plot (of target - SalePrice in this case) of imputed and original values does not overlap -- bad thing
# Cuz missing data was a lot -- after imputation the distribution changes a lot

# FOR GarageQual
# After imputation the KDE plot (of target - SalePrice in this case) of original data and imputed data remains same -- Good thing
# Why? -- Cuz missing values were <5% 
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))

sns.barplot(data=df['FireplaceQu'].value_counts(), ax=ax[0], label='Before Imputation', alpha=0.7, color='black')
sns.barplot(data=df['FireplaceQu_imputed'].value_counts(), ax=ax[0], label='After Imputation', alpha=0.7)

sns.barplot(data=df['GarageQual'].value_counts(), ax=ax[1], label='Before Imputation', alpha=0.7, color='black')
sns.barplot(data=df['GarageQual_imputed'].value_counts(), ax=ax[1], label='After Imputation', alpha=0.7)

ax[0].legend()
ax[1].legend()
plt.suptitle('Distribution of values before and after imputation')
plt.show()
```

```python
# From the barplot also we can see that the FireplaceQu mode changes a lot after imputation
```
