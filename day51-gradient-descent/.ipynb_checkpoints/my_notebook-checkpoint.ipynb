```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=10, n_features=1, n_targets=1, noise=10, random_state=95)
```

```python
X = X.ravel()
```

```python
sns.scatterplot(x=X, y=y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

lr = LinearRegression()
lr.fit(X_train.reshape(-1,1), y_train)

lr.coef_, lr.intercept_
```

```python
X_plot = np.linspace(-2, 2, 10)
y_plot = lr.predict(X_plot.reshape(-1,1))

sns.scatterplot(x=X, y=y)
sns.lineplot(x=X_plot, y=y_plot)
```

```python
# y = mx + b

m = 62.63
b_original = -4.48

X=np.linspace(-2,2,10)
y=m*X+b_original

b = 500
lr = 0.02

y_set = m*X+b

# iteration 1
loss_slope = -2 * sum(y-(m*X)-b)
b = b - (lr * loss_slope)
y_new = m*X+b

sns.lineplot(x=X, y=y, label=f'b={b_original}')
sns.lineplot(x=X, y=y_set, color='red', label=f'b={b}')
sns.lineplot(x=X, y=y_new, color='orange', label=f'b={b}')
```

```python
# iteration 2
loss_slope = -2 * sum(y-(m*X)-b)
b = b - (lr * loss_slope)
y_new_2 = m*X+b

sns.lineplot(x=X, y=y, label=f'b={b_original}', color='black')
sns.lineplot(x=X, y=y_set, color='red')
sns.lineplot(x=X, y=y_new, color='orange')
sns.lineplot(x=X, y=y_new_2, color='yellow')
```

```python
# iteration 3
loss_slope = -2 * sum(y-(m*X)-b)
b = b - (lr * loss_slope)
y_new_3 = m*X+b

sns.lineplot(x=X, y=y, label=f'b={b_original}', color='black')
sns.lineplot(x=X, y=y_set, color='red')
sns.lineplot(x=X, y=y_new, color='orange')
sns.lineplot(x=X, y=y_new_2, color='yellow')
sns.lineplot(x=X, y=y_new_3, color='green')
```

```python
# iteration 4
loss_slope = -2 * sum(y-(m*X)-b)
b = b - (lr * loss_slope)
y_new_4 = m*X+b

sns.lineplot(x=X, y=y, label=f'b={b_original}', color='black')
sns.lineplot(x=X, y=y_set, color='red')
sns.lineplot(x=X, y=y_new, color='orange')
sns.lineplot(x=X, y=y_new_2, color='yellow')
sns.lineplot(x=X, y=y_new_3, color='green')
sns.lineplot(x=X, y=y_new_4, color='blue')
```

```python
# iteration 5
loss_slope = -2 * sum(y-(m*X)-b)
b = b - (lr * loss_slope)
y_new_5 = m*X+b

sns.lineplot(x=X, y=y, label=f'b={b_original}', color='black')
sns.lineplot(x=X, y=y_set, color='red')
sns.lineplot(x=X, y=y_new, color='orange')
sns.lineplot(x=X, y=y_new_2, color='yellow')
sns.lineplot(x=X, y=y_new_3, color='green')
sns.lineplot(x=X, y=y_new_4, color='blue')
sns.lineplot(x=X, y=y_new_5, color='violet')
```

```python
plt.figure(figsize=(20,15))

m = 62.63
b_OLS = -4.48

X = np.linspace(-2,2,10)
y = m*X + b_OLS

sns.lineplot(x=X, y=y, label=f'b_OLS={b_OLS}')

b = 500
y_pred = m*X + b
sns.lineplot(x=X, y=y_pred, label=f'b={b}')

lr=0.01
epochs = 50
for epoch in range(epochs):    
    
    loss_slope = -2 * sum(y-(m*X)-b)
    b = b - (lr * loss_slope)
    y_pred = m*X+b
    sns.lineplot(x=X, y=y_pred, label=f'b={np.round(b,2)}')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.legend(loc='right')
    
    print(f'Epoch {epoch} : b = {b}')
```

```python
plt.figure(figsize=(20,15))

m = 62.63
b_OLS = -4.48

X = np.linspace(-2,2,10)
y = m*X + b_OLS

sns.lineplot(x=X, y=y, label=f'b_OLS={b_OLS}')

b = 500
y_pred = m*X + b
sns.lineplot(x=X, y=y_pred, label=f'b={b}')

lr=0.01
epoch = 0

while 1:    
    
    loss_slope = -2 * sum(y-(m*X)-b)
    
    # adding a stopping condition
    if lr*loss_slope < 0.001:
        break
        
    epoch +=1
    b = b - (lr * loss_slope)
    y_pred = m*X+b
    sns.lineplot(x=X, y=y_pred, label=f'b={np.round(b,2)}')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.legend(loc='right')
    
    print(f'Epoch {epoch} : b = {b}')
```
